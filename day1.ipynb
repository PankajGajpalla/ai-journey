````python
class AIAgent:
    def __init__(self, name, model, max_tokens):
        self.name = name
        self.model = model
        self.max_tokens = max_tokens
        self.messages = []
        self.turn_count = 0
    
    def add_message(self, role, content):
        self.messages.append({"role": role, "content": content})
    
    def get_history(self):
        for m in self.messages:
            print(f"user: {m['role']}")
            print(f"assistant: {m['content']}")
    
class ResearchAgent(AIAgent):
    def __init__(self, name, model, max_tokens, sources):
        super().__init__(name, model, max_tokens)
        self.sources = sources

    def search(self, query):
        print(f"Searching {self.sources} for: {query}")
        self.add_message("tool", f"searched for {query}")

class WriterAgent(AIAgent):
    def __init__(self, name, model, max_tokens, style):
        super().__init__(name, model, max_tokens)
        self.style = style
    
    def write(self, topic):
        print(topic)
        self.add_message("tool", f"wrote about {topic}")


researcher = ResearchAgent("ResearchBot", "claude-sonnet-4-6", 1000, ["web", "arxiv"])
writer = WriterAgent("WriterBot", "claude-sonnet-4-6", 1000, "professional")

researcher.add_message("user", "Find papers on AI agents")
researcher.search("AI agents")
researcher.add_message("assistant", "Found 5 relevant papers")

writer.add_message("user", "Write a blog post about AI")
writer.write("AI agents")
writer.add_message("assistant", "Blog post complete")

researcher.get_history()
print("---")
writer.get_history()
print("---")
print(researcher)
print(writer)
print(f"Researcher turns: {researcher.turn_count}")
````

````python
class AIAgent:
    def __init__(self, name, model, max_tokens):
        self.name = name
        self.model = model
        self.max_tokens = max_tokens
        self.messages = []
        self.turn_count = 0

    def add_message(self, role, content):
        self.messages.append({"role": role, "content": content})
        self.turn_count += 1

    def get_history(self):
        for m in self.messages:
            print(f"{m['role']}: {m['content']}")

    def reset(self):
        self.messages = []
        self.turn_count = 0

    def __str__(self):
        return f"AIAgent(name={self.name}, turns={self.turn_count})"


class ResearchAgent(AIAgent):
    def __init__(self, name, model, max_tokens, sources):
        super().__init__(name, model, max_tokens)
        self.sources = sources

    def search(self, query):
        print(f"Searching {self.sources} for: {query}")
        self.add_message("tool", f"searched for {query}")


class WriterAgent(AIAgent):
    def __init__(self, name, model, max_tokens, style):
        super().__init__(name, model, max_tokens)
        self.style = style

    def write(self, topic):
        print(f"Writing about {topic} in {self.style} style")
        self.add_message("tool", f"wrote about {topic}")


researcher = ResearchAgent("ResearchBot", "claude-sonnet-4-6", 1000, ["web", "arxiv"])
writer = WriterAgent("WriterBot", "claude-sonnet-4-6", 1000, "professional")

researcher.add_message("user", "Find papers on AI agents")
researcher.search("AI agents")
researcher.add_message("assistant", "Found 5 relevant papers")

writer.add_message("user", "Write a blog post about AI")
writer.write("AI agents")
writer.add_message("assistant", "Blog post complete")

researcher.get_history()
print("---")
writer.get_history()
print("---")
print(researcher)
print(writer)
print(f"Researcher turns: {researcher.turn_count}")
````

Part 1 — Decorators
Write two decorators:
@logger — prints "Calling: function_name" before and "Done: function_name" after any function runs
@validate_input — checks if all arguments passed to a function are not empty strings or None. If any argument is invalid, prints "Error: Invalid input" and returns None without running the function
Apply both decorators to this function:
python@logger
@validate_input
def create_agent(name, model):
    return f"Agent {name} created with {model}"
Test with:
pythonprint(create_agent("ResearchBot", "claude-sonnet-4-6"))  # should work
print(create_agent("", "claude-sonnet-4-6"))              # should catch empty string
print(create_agent("ResearchBot", None))                  # should catch None


````python
def logger(func):
    def wrapper(*args, **kwargs):
        print(f"Calling: {func.__name__}")
        result = func(*args, **kwargs)
        print(f"Done: {func.__name__}")
        return result
    return wrapper


def validate_input(func):
    def wrapper(*args, **kwargs):
        for a in args:
            if a == "" or a is None:
                print("Error: Invalid input")
                return None
        return func(*args, **kwargs)
    return wrapper


@logger
@validate_input
def create_agent(name, model):
    return f"Agent {name} created with {model}"


print(create_agent("ResearchBot", "claude-sonnet-4-6"))
print(create_agent("", "claude-sonnet-4-6"))
print(create_agent("ResearchBot", None))


def message_streamer(messages):
    for m in messages:
        if m['role'] == 'user' or m['role'] == 'assistant':
            yield f"[{m['role']}] {m['content']}"


history = [
    {"role": "user", "content": "Find papers on agents"},
    {"role": "tool", "content": "searched web"},
    {"role": "assistant", "content": "Found 5 papers"},
    {"role": "user", "content": "Summarize them"},
    {"role": "tool", "content": "processing"},
    {"role": "assistant", "content": "Here is the summary"},
]

for msg in message_streamer(history):
    print(msg)
````

````python
import csv
from datetime import datetime
import json

class AgentLogger:
    def __init__(self, log_file, csv_file):
        self.log_file = log_file
        self.csv_file = csv_file
        self.logs = []

    def log_run(self, agent_name, task, tokens_used, success):
        dic = {"agent_name":agent_name, "task":task, "tokens_used":tokens_used, "success":success}
        dic["timestamp"] = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        self.logs.append(dic)
    
    def save_to_csv(self):
        with open(self.csv_file, "w") as file:
            writer = csv.DictWriter(file, fieldnames=["agent_name", "task", "tokens_used", "success", "timestamp"])
            writer.writeheader()
            writer.writerows(self.logs)
    
    def save_to_json(self):
        with open(self.log_file, "w") as file:
            json.dump(self.logs, file)
    
    def get_successful_runs(self):
        return [log for log in self.logs if log["success"]]

    def get_total_tokens_used(self):
        s = 0
        for a in self.logs:
            s += a["tokens_used"]

        return s
    

    def get_summary(self):
        print(f"Total runs: {len(self.logs)}")
        print(f"Successful: {len(self.get_successful_runs())}")
        print(f"Failed: {len(self.logs)-len(self.get_successful_runs())}")
        print(f"Total tokens used: {self.get_total_tokens_used()}")
        print(f"Most recent: {self.logs[-1]['timestamp']}")
    

logger = AgentLogger("logs.json", "logs.csv")

logger.log_run("ResearchBot", "Find AI papers", 1200, True)
logger.log_run("WriterBot", "Write blog post", 800, True)
logger.log_run("ReviewBot", "Review content", 600, False)
logger.log_run("ResearchBot", "Find ML papers", 1100, True)
logger.log_run("WriterBot", "Write summary", 800, True)

logger.save_to_csv()
logger.save_to_json()
logger.get_summary()

successful = logger.get_successful_runs()
print(f"\nSuccessful runs: {len(successful)}")
for run in successful:
    print(f"  {run['agent_name']} — {run['task']}")

        
````

````python
import csv
import json
from datetime import datetime


class AgentLogger:
    def __init__(self, log_file, csv_file):
        self.log_file = log_file
        self.csv_file = csv_file
        self.logs = []

    def log_run(self, agent_name, task, tokens_used, success):
        entry = {
            "agent_name": agent_name,
            "task": task,
            "tokens_used": tokens_used,
            "success": success,
            "timestamp": datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        }
        self.logs.append(entry)

    def save_to_csv(self):
        with open(self.csv_file, "w", newline="") as file:
            writer = csv.DictWriter(file, fieldnames=["agent_name", "task", "tokens_used", "success", "timestamp"])
            writer.writeheader()
            writer.writerows(self.logs)

    def save_to_json(self):
        with open(self.log_file, "w") as file:
            json.dump(self.logs, file, indent=4)

    def get_successful_runs(self):
        return [log for log in self.logs if log["success"]]

    def get_total_tokens_used(self):
        return sum(log["tokens_used"] for log in self.logs)

    def get_summary(self):
        print(f"Total runs: {len(self.logs)}")
        print(f"Successful: {len(self.get_successful_runs())}")
        print(f"Failed: {len(self.logs) - len(self.get_successful_runs())}")
        print(f"Total tokens used: {self.get_total_tokens_used()}")
        print(f"Most recent: {self.logs[-1]['timestamp']}")


logger = AgentLogger("logs.json", "logs.csv")

logger.log_run("ResearchBot", "Find AI papers", 1200, True)
logger.log_run("WriterBot", "Write blog post", 800, True)
logger.log_run("ReviewBot", "Review content", 600, False)
logger.log_run("ResearchBot", "Find ML papers", 1100, True)
logger.log_run("WriterBot", "Write summary", 800, True)

logger.save_to_csv()
logger.save_to_json()
logger.get_summary()

successful = logger.get_successful_runs()
print(f"\nSuccessful runs: {len(successful)}")
for run in successful:
    print(f"  {run['agent_name']} — {run['task']}")
````

## HTTP & REST APIs

Every AI agent you build will talk to the outside world — searching the web, calling Claude's API, fetching data, sending results. All of that happens through HTTP. Understanding this is non-negotiable.

---

## What is HTTP?

HTTP is the language computers use to talk to each other over the internet. When you open a website, your browser sends an HTTP **request** to a server. The server sends back an HTTP **response**.

```
Your Python code                    Server
      │                                │
      │──── HTTP Request ────────────►│
      │                                │ (processes request)
      │◄─── HTTP Response ────────────│
      │                                │
```

That's it. Request and response. Everything on the internet works this way.

---

## HTTP Methods

Every request has a method — what you want to do:

```
GET     → fetch data          (read)
POST    → send data           (create)
PUT     → replace data        (update)
PATCH   → partially update    (update part)
DELETE  → delete data         (delete)
```

As an AI engineer you'll mostly use `GET` and `POST`.

- `GET` — "give me data" — fetching search results, getting weather, reading a page
- `POST` — "here's data, do something with it" — sending a prompt to Claude's API, submitting a form

---

## HTTP Status Codes

Every response comes with a number telling you what happened:

```
200 → OK — worked perfectly
201 → Created — new resource created
400 → Bad Request — your request had an error
401 → Unauthorized — wrong or missing API key
403 → Forbidden — you don't have permission
404 → Not Found — that resource doesn't exist
429 → Too Many Requests — you're being rate limited
500 → Internal Server Error — server broke
```

You'll check these constantly in agent code. A `429` means slow down. A `401` means check your API key. A `500` means the server is down — retry later.

---

## What is a REST API?

A REST API is a server that follows standard rules for how URLs and HTTP methods work together. Each URL represents a resource:

```
GET  https://api.example.com/agents          → get all agents
GET  https://api.example.com/agents/123      → get agent with id 123
POST https://api.example.com/agents          → create a new agent
DELETE https://api.example.com/agents/123    → delete agent 123
```

Clean, predictable, consistent. That's REST.

---

## JSON — the language of APIs

Almost every API sends and receives data as JSON. You already know JSON from file handling. Now you'll see it flying over the internet:

```json
{
    "id": "msg_123",
    "model": "claude-sonnet-4-6",
    "content": [
        {
            "type": "text",
            "text": "Python is a programming language"
        }
    ],
    "usage": {
        "input_tokens": 10,
        "output_tokens": 8
    }
}
```

That's an actual Anthropic API response. Your Python code receives this, parses it, and extracts what it needs.

## Headers — sending extra information

Headers are key-value pairs sent with every request. They carry metadata — who you are, what format you want, your API key:

```python
import requests

headers = {
    "Content-Type": "application/json",
    "User-Agent": "MyAgentApp/1.0"
}

response = requests.get(
    "https://official-joke-api.appspot.com/random_joke",
    headers=headers
)

print(response.status_code)
```

The most important header you'll use is `Authorization` — this is how you send API keys:

```python
headers = {
    "Authorization": "Bearer your-api-key-here",
    "Content-Type": "application/json"
}
```

Every call to Claude's API uses exactly this pattern.

## Error handling for API calls

Networks fail. APIs go down. Always wrap API calls in error handling:

```python
import requests

def fetch_joke():
    try:
        response = requests.get(
            "https://official-joke-api.appspot.com/random_joke",
            timeout=5   # fail after 5 seconds if no response
        )
        response.raise_for_status()   # raises exception for 4xx/5xx status codes
        return response.json()

    except requests.exceptions.Timeout:
        print("Request timed out")
        return None
    except requests.exceptions.ConnectionError:
        print("No internet connection")
        return None
    except requests.exceptions.HTTPError as e:
        print(f"HTTP error: {e}")
        return None
    except Exception as e:
        print(f"Unexpected error: {e}")
        return None

joke = fetch_joke()
if joke:
    print(joke["setup"])
    print(joke["punchline"])
```

`timeout=5` — never let a request hang forever  
`raise_for_status()` — auto-raises exception for bad status codes  

This pattern is exactly what you'll use for every AI API call.

## Environment variables — storing API keys safely

Never put API keys directly in your code:

```python
# ❌ NEVER do this — if you push to GitHub, your key is exposed
api_key = "sk-ant-api03-real-key-here"
```

Always use environment variables:

**Step 1 — create a `.env` file in your project folder:**
```
ANTHROPIC_API_KEY=sk-ant-api03-your-key-here
OPENAI_API_KEY=sk-your-key-here
```

**Step 2 — install python-dotenv:**
```bash
pip install python-dotenv
```

**Step 3 — load in your code:**
```python
from dotenv import load_dotenv
import os

load_dotenv()   # reads .env file and loads variables

api_key = os.getenv("ANTHROPIC_API_KEY")
print(api_key)   # → your key, safely loaded
```

**Step 4 — add `.env` to `.gitignore`:**
```
.env
venv/
__pycache__/
```

Your key never touches GitHub. This is standard practice everywhere.

````python
import requests
import json


def fetch_joke(category):
    try:
        response = requests.get(
            f"https://official-joke-api.appspot.com/jokes/{category}/random",
            timeout=5
        )
        response.raise_for_status()
        data = response.json()
        return data[0] if isinstance(data, list) else data
    
    except requests.exceptions.Timeout:
        print("Reqest timed out")
        return None
    except requests.exceptions.ConnectionError:
        print("No internet")
        return None
    except requests.exceptions.HTTPError as e:
        print(f"HTTP error: {e}")
        return None
    except Exception as e:
        print(f"Unexpected error: {e}")
        return None
    
def display_joke(joke):
    print(f"Category: {joke['type']}")
    print(f"Q: {joke['setup']}")
    print(f"A: {joke['punchline']}")

class JokeFetcher:
    def __init__(self, category):
        self.category =  category
        self.history = []

    def fetch(self):
        jk = fetch_joke(self.category)
        if jk:
            self.history.append(jk)
        return jk

    def get_history(self):
        for i,j in enumerate(self.history):
            print(f"{i+1}")
            display_joke(j)
            print()
    
    def save_history(self, filename):
        with open(filename, "w") as file:
            json.dump(self.history, file)



fetcher = JokeFetcher("programming")

for i in range(3):
    joke = fetcher.fetch()
    if joke:
        display_joke(joke)
        print()

fetcher.save_history("joke_history.json")
print(f"Total fetched: {len(fetcher.history)}")
            

    
````

````python
import asyncio
import time

async def fetch_web_search(query):
    await asyncio.sleep(2)   # simulating 2 second API call
    return f"Web results for {query}"

async def fetch_news(query):
    await asyncio.sleep(2)
    return f"News results for {query}"

async def fetch_academic(query):
    await asyncio.sleep(2)
    return f"Academic results for {query}"

async def main():
    start = time.time()

    results = await asyncio.gather(
        fetch_web_search("AI agents"),
        fetch_news("AI agents"),
        fetch_academic("AI agents")
    )

    print(f"Total time: {time.time() - start:.1f}s")
    for r in results:
        print(r)

asyncio.run(main())
# → Total time: 2.0s
````

````python
import aiohttp
import asyncio


async def fetch_joke(session, category):
    url = f"https://official-joke-api.appspot.com/jokes/{category}/random"
    try:
        async with session.get(url, timeout=aiohttp.ClientTimeout(total=5)) as response:
            response.raise_for_status()
            data = await response.json()
            return data[0] if isinstance(data, list) else data
    except aiohttp.ClientConnectionError:
        print("No internet")
        return None
    except aiohttp.ClientResponseError as e:
        print(f"HTTP Error: {e}")
        return None
    except Exception as e:
        print(f"Unexpected error: {e}")
        return None


async def fetch_all_jokes(categories):
    async with aiohttp.ClientSession() as session:
        jokes = await asyncio.gather(*[
            fetch_joke(session, cat) for cat in categories
        ])
        return jokes


def summarize(jokes):
    valid = [j for j in jokes if j is not None]
    print(f"Fetched {len(valid)} jokes:\n")
    for i, joke in enumerate(valid):
        print(f"{i+1}. [{joke['type']}] {joke['setup']} — {joke['punchline']}")


async def main():
    categories = ["programming", "general", "knock-knock"]
    jokes = await fetch_all_jokes(categories)
    summarize(jokes)


asyncio.run(main())
````
